```python
import os
%matplotlib inline
import matplotlib.pyplot as plt

path_train = './dataset_birds/train/'

folders = []
for item in os.listdir(path_train):
    folders.append(item)
```

```python
images_number = []
```

```python
for folder in folders:
    items = os.listdir(path_train+folder)
    images_number.append(len(items))
```

```python
images_number
```

```python
plt.figure(figsize = (25,15))
plt.plot(folders, images_number)
plt.xticks([]);

```
